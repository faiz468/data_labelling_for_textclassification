```python
label = ["Medical Assistance", "Security", "Divyangjan Facilities", "Facilities for Women with Special needs",
         "Electrical Equipment", "Coach - Cleanliness", "Punctuality", "Water Availability",
         "Coach - Maintenance", "Catering & Vending Services", "Staff Behaviour",
         "Corruption / Bribery", "Bed Roll", "Miscellaneous"]

```

```python
l=[]
i=1700
while(i<1800):
    l.append(classifier(df['tweet'][i], label))
    i+=1
```

```python
l
```

```python
l1=[]
for i in range(len(l)):
    max = 0
    for j in range(len(l[i]['scores'])):
        if l[i]['scores'][j]>l[i]['scores'][max]:
            max = j
    l1.append({'tweet':f"{l[i]['sequence']}", 'label':f"{l[i]['labels'][max]}", 'score':f"{l[i]['scores'][max]}"})
```

```python
dff = pd.DataFrame(l1)
```
